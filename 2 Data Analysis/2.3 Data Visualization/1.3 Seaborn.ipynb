#### Seaborn

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a higher-level, more elegant, and easier-to-use-interface for creating visually appealing and informative statistical graphics.

In simple terms:

> Matplotlib gives you full control but is low-level (you must specify everything manually).

> Seaborn simplies the process by providing ready-to-use functions for common statistical plots with beautiful default styles.

````python
# Example
import seaborn as sns
import matplotlib.pyplot as plt

# sample data
tips = sns.load_dataset('tips')

# simple seaborn plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()
````

##### Line Plot

##### Syntax:

```Python
sns.lineplot(x=None, 
            y=None,
            hue=None, 
            style=None,
            size=None,           
            data=None, 
            estimator='mean',
            errorbar=('ci', 95), 
            markers=False, 
            dashes=True)
```
---
##### Parameters

| Parameter     | Description                                                                             |
| ------------- | --------------------------------------------------------------------------------------- |
| **x, y**      | Variables to plot on x and y axes.                                                      |
| **hue**       | Categorical variable that defines line color (e.g., gender, species).                   |
| **style**     | Changes the line style based on a variable (e.g., dashed for one group).                |
| **size**      | Varies the thickness of lines by a variable.                                            |
| **data**      | Dataset (usually a Pandas DataFrame).                                                   |
| **estimator** | Statistical function applied to y-values (default `'mean'`). Use `None` for raw values. |
| **errorbar**  | Type of error bar or confidence interval (`'ci'`, `'sd'`, or `None`).                   |
| **markers**   | If `True`, draws a marker at each data point.                                           |
| **dashes**    | If `True`, enables dashed lines for style variations.                                   |
| **palette**   | Defines the color palette for multiple lines.                                           |
| **ax**        | Matplotlib axes to draw the plot on.                                                    |
---
##### Illustrated Idea

Dataset: Average daily temperature across three cities over 12 months.

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#* Example Dataset
data = pd.DataFrame({
    "month": list(range(1, 13)) * 3,
    "temperature": [25,27,30,33,35,36,34,32,30,28,26,25,
                    10,12,15,18,22,25,24,20,18,14,12,10,
                    15,17,20,24,28,30,29,26,24,20,18,16],
    "city": ["Mumbai"]*12 + ["Delhi"]*12 + ["Bangalore"]*12
})

sns.lineplot(x='month', y='temperature', hue='city', marker='o', data=data)
plt.title('Monthly Temperature Trends Across Cities')
plt.xlabel('Month')
plt.ylabel('Temparature (c)')
plt.show()
````

##### Key Functions
- Function:`sns.lineplot()`

- Best for: Trends, time series, or continuous relationships

- Supports confidence intervals and grouped summaries out of the box.

- Handles Pandas DataFrames naturally.

- Easier and cleaner alternative to `plt.plot()` when working with structured data.

- Use `estimator=None` to show all raw data points rather than means.

- Combine with Seaborns themes (`sns.set_theme(style='whitegrid')`) for publication-ready visuals.

#### Box Plot

##### How it works?

`sns.boxplot()`:

1. Calculates Q1, median (Q2), and Q3 from the data.

2. The box spans Q1 to Q3 the (IQR).

3. The line inside the box represents the median.

4. The whiskers extend to data points withing 1.5xIQR from Q1 to Q3.

5. Points beyond whiskers are shown as outliers (dots).

6. When given a categorical variable, it creates multiple boxes for comparison.
---
##### Syntax

```Python
sns.boxplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=.75, width=0.8, dodge=True, fliersize=5, linewidth=None, whis=1.5)
```
---
##### Parameters

| Parameter             | Description                                                  |
| --------------------- | ------------------------------------------------------------ |
| **x, y**              | Variables to plot (x = category, y = numeric).               |
| **hue**               | Adds a second categorical variable for color-coded grouping. |
| **data**              | Dataset (Pandas DataFrame).                                  |
| **order / hue_order** | Custom category order.                                       |
| **palette**           | Color scheme (`"pastel"`, `"Set2"`, `"coolwarm"`, etc.).     |
| **orient**            | `'v'` for vertical, `'h'` for horizontal.                    |
| **width**             | Width of each box (default 0.8).                             |
| **dodge**             | Whether to separate boxes for hue categories.                |
| **fliersize**         | Size of outlier markers.                                     |
| **linewidth**         | Thickness of box edges.                                      |
| **whis**              | Controls whisker length (default = 1.5  IQR).               |
| **saturation**        | Controls color intensity.                                    |
---
##### Illustration Idea

Example 1: Basic Box Plot

````python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.boxplot(x='day', y='total_bill', hue='day', legend=False, data=tips, palette='Set3')

plt.title('Distribution of total bills by day')
plt.xlabel('Day of the week')
plt.ylabel('Total Bill ($)')
plt.show()
````

````python
sns.boxplot(x='smoker', y='total_bill', hue='smoker', legend=False, data=tips, whis=2.0, fliersize=4, palette='pastel')

plt.title('Total Bill Distribution (Custom Whiskers and Outliers)')
plt.show()
````

---
##### Key Points

- Function: `sns.boxplot()`

- Best for: Comparing distributions and detecting outliers.

- shows median, whiskers, IQR, and outliers.

- Add `hue` for grouped comparisons.

- Great for feature comparison, data quality checks, and outlier detection.

- Works seamlessly with Pandas DataFrame.

#### Pair Plot

##### What is it?

A Pair Plot in Seaborn (created using `sns.pairplot()`) is a grid of scatter plots and histograms that shows pairwise relationships between multiple numerical variables in a dataset.

Each cell in the grid represents the relationship between two variables, while the diagonal shows the distribution (usually histogram or KDE plots) of each variable.

Essentially, it's compact visual summary of how every variable rlates to every other variable -- ideal for exploratory data analysis.

---
##### Why use it?

- To explore relationships (correlations, trends, clusters) between variables.

- To detect patterns, outliers, and non-linear relationships.

- To visually identify potential predictors or redundant variables.

- To understand feature behavior before machine learning modeling.

- To compare subgroups using the hue parameter (e.g., male vs female, smoker vs non-smoker).

Example uses:

- Exploring relationships among features in the Iris, Tips, or Titanic datasets.

- Checking for linear correlation between variables like price, area, and rooms in housing data.

- Identifying class separability before classification tasks.
---
##### How it works?

`sns.pairplot()`

1. Takes a dataseet (usually a pandas dataframe)

2. Plots scatter plots between every pair of numerical variables.

3. Draws histograms or KDE plots on the diagonal to show distributions.

4. Optionally applies a categorical hue to color-code data points by class or category.

5. Allows customization of markers, sized, and transperancy for clarity.
---
##### Syntax

```Python
sns.pairplot(data, vars=None, hue=None, kind='scatter', diag_kind='auto', palette=None, markers=None, corner=False, height=2.5, plot_kws=None)
```
---
##### Parameters

| Parameter     | Description                                                                    |
| ------------- | ------------------------------------------------------------------------------ |
| **data**      | The dataset (Pandas DataFrame).                                                |
| **vars**      | Subset of variables to plot (list of column names).                            |
| **hue**       | Adds color coding based on a categorical variable.                             |
| **kind**      | Type of plot for off-diagonal cells (`'scatter'`, `'kde'`, `'hist'`, `'reg'`). |
| **diag_kind** | Plot type for diagonal (`'auto'`, `'hist'`, `'kde'`, `'None'`).                |
| **palette**   | Color palette for hue (e.g., `'pastel'`, `'coolwarm'`).                        |
| **markers**   | Marker styles for different hue levels.                                        |
| **corner**    | If `True`, plots only the lower triangle (to save space).                      |
| **height**    | Size (in inches) of each subplot.                                              |
| **plot_kws**  | Additional keyword arguments for the plots (like alpha, s, linewidth).         |
---
##### Illustration Idea

Example 1: Basic Pair Plot

````python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

sns.pairplot(iris)

plt.suptitle('Pairwise relationship in the Iris Dataset' , y=1.02)
plt.show()
````

**Explanation:**

Each pair of numerical features (sepal length, sepal width, petal length, petal width) is plotted against the others.

The diagonal shows each variables distribution.

Its ideal for spotting strong relationships (like between petal length and width).

Example 2: Pair Plot with Hue

````python
sns.pairplot(iris, hue='species', palette='Set2')

plt.suptitle('Iris Dataset Pair Plot by Species', y=1.02)
plt.show()
````

**Explanation:**

This color-codes data points by species (`setosa`,`versicolor`, `verginica`).

You can clearly see how the three species seperate -- perfect classification insights.

Example 3: KDE and Corner Plot

````python
sns.pairplot(iris, kind='kde', diag_kind='kde', hue='species', corner=True, palette='pastel')

plt.suptitle('Smoothed Pair Plot (KDE) - Iris Dataset', y=1.02)
plt.show()
````

**Explanation:**

- Uses KDE (Kernel Density Estimation) instead of scatter plots for smoother contours.

- `corner=True` reduces redudancy by showing only the lower half of the grid.

- Excellent for dense datasets where scatter points would overlap.

Example 4: Customizing with plot_kws

````python
sns.pairplot(iris, hue='species', plot_kws={'alpha':0.7, 's':60, 'edgecolor':'w'})

plt.suptitle('Customized Pair Plot with Transperancy', y=1.02)
plt.show()
````

**Explanation**

We can fine-tune the visuals -- transparency (`alpha`), size(`s`), and edge color improve readibility in crowded plots.

##### Key Points

- Function: `sns.pairplot()`

- Best for: Exploring pairwise relationships between numerical variables.

- Automatically generates scatter + histogram/KDE plots.

- Use `hue` to reveal class differences visually.

- Use `corner=True` for compact grids.

- Works great for correlation detection, clustering analysis, and feature selection.

- For large datasets, set `plot_kws={'alpha':0.5}` to avoid overlap.

- Can be computationally heavy on very large datasets  use subsets (`vars=[...]`).

#### Heatmap

##### What is Heatmap?

A heatmap is a color-coded representation of data where values are represented by different colors.

It's most commonly used to visualize matrices, such as correlation matrices, confusion metrices, or any 2D tabular data.

Each cell's color intensity correspons to the magnitude of the value -- making patterns, trends, and relatioships instantly visible.

---
##### Why use Heatmap?

- To visualize correlations between numerical variables.

- To explore patterns or anomalies in 2D data (e.g., high or low values).

- To display confusion matrices in classification.

- To summarize large amounts of data visually.

- To quickly spot clusters or relationships.
---
##### Syntax

```Python 
sns.heatmap(data, annot=None, fmt=None, cmap=None, cbar=True, linewidth=0, linecolor='white', vmin=None, vmax=None, center=None, square=False, mask=None, ax=None)
```
---
##### Parameters

| Parameter          | Description                                                         |
| ------------------ | ------------------------------------------------------------------- |
| **data**           | The dataset (2D array, DataFrame, or matrix).                       |
| **annot**          | If `True`, writes the value in each cell.                           |
| **fmt**            | String formatting code (like `'.2f'` for two decimals).             |
| **cmap**           | Colormap (e.g., `'coolwarm'`, `'viridis'`, `'plasma'`, `'YlGnBu'`). |
| **cbar**           | Displays a color bar (True/False).                                  |
| **linewidths**     | Width of lines dividing cells.                                      |
| **linecolor**      | Color of dividing lines.                                            |
| **vmin**, **vmax** | Control color scaling range.                                        |
| **center**         | Set midpoint of colormap (e.g., `center=0` for correlation).        |
| **square**         | Force square cells.                                                 |
| **mask**           | Mask (hide) parts of the heatmap.                                   |
---
##### Illustration Idea

Example 1: Correlation Heatmap

````python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

corr = iris.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Heatmap - Iris Dataset')
plt.show()
````

**Explanation:**

- `iris.corr()` calculates correlations between numerical columns.

- `annot=True` displays correlation coefficients in each cell.

- The color scheme clearly displays positive and negetive correlations.

Example 2: Masking Upper Triangle (Cleaner Look)

````python
import numpy as np

mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(corr, mask=mask, annot=True, cmap='YlGnBu', square=True, fmt='.2f')

plt.title('Lower-Triangle Correlation Heatmap - Iris Dataset')
plt.show()
````

**Explanation**

- `mask=np.triu(...)` hides the upper triangle since correlation metrices are symmetric.

- `square=True` gives uniform square cells for neatness.

Example 3: Heatmap from custom Dataset

````python
data = [[12, 19, 3],
        [5, 10, 15],
        [7, 8, 9]]

sns.heatmap(data, annot=True, cmap='magma', linewidths=1, linecolor='white')

plt.title('Basic 3x3 Heatmap')
plt.show()
````

Example 4: Confusion Matrix Heatmap (ML Example)

````python
import pandas as pd
from sklearn.metrics import confusion_matrix

y_true = [1, 0, 1, 1, 0, 1, 0]
y_pred = [1, 0, 1, 0, 0, 1, 1]

cm = confusion_matrix(y_true, y_pred)
labels = ['Negetive', 'Positive']

sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix Heatmap')
plt.show()
````

**Explanation**

- Common use in machine learning to visualize model accuracy and misclassifications.

Example 5 : Centered Diverging Heatmap

````python
sns.heatmap(corr, annot=True, cmap='RdBu_r', center=0)
plt.title('Centered Diverging Heatmap')
plt.show()
````

**Explanation**

- Setting `center=0` makes positive and negetive correlations diverge symmetrically -- great for identifying polarit in correlations.

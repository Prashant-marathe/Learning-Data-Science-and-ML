#### Line Plot

#### What is it?

A line plot (or line chart) is one of the most basic and widely used plots in Matplotlib. It displays data as a series of points connected by straight lines. The x-axis typically represents independent variable (like time or sequence), and the y-axis shows a dependent variable (like measurement values). It's ideal for showing trends, changes, or relationships over a continuous range.

---
##### Why use it?

Line plots are useful for:

- Showing trends over time (e.g stock prices, temperature changes, or performance matrices).

- Comparing multiple datasets across the same x-axis.

- Highlighting patterns, fluctuations, or continuity in data.

- Quickly understanding how one variable shaves with respect to another.

---
##### How it works?

A line plot connects a sequence of (x, y) data points with lines.

Matplotlib takes arrays or losts of x and y values, and draws lines sequentially from one point to the next. You can customize line style, color, width, and markers to differentiate between multiple datasets or highlight specific trends.

---
##### Syntax

```Python
import matplotlib.pyplot as plt

plt.plot(x, y, color='blue', linestyle='-.', marker='o')
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Line plot Example')
plt.show()
```
---
##### Parameters

| Parameter   | Description                                                     |
| ----------- | --------------------------------------------------------------- |
| `x, y`      | Data points to plot (arrays, lists, or Series).                 |
| `color`     | Color of the line (e.g., `'r'` for red, `'blue'`, `'#123456'`). |
| `linestyle` | Style of the line (`'-'`, `'--'`, `'-.'`, `':'`).               |
| `marker`    | Symbol to mark data points (`'o'`, `'*'`, `'s'`, etc.).         |
| `linewidth` | Thickness of the line.                                          |
| `label`     | Label for the line (used with `plt.legend()`).                  |
| `alpha`     | Transparency level (0 = transparent, 1 = opaque).               |
---

##### Illustration Idea

You could plot the daily temperature of a ciry over a week:

````python
days = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
temparature = [30, 32, 33, 31, 29, 28, 27]

plt.plot(days, temparature, marker='o', color='green')
plt.title('Temperature over a week')
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.grid(True)
plt.show()
````

##### Key Points

- Best for sequential and continuous data.

- Use different colors or line styles to compare multiple lines.

- Combine with markers to emphasize specific points.

- Keep the plot simple and clean --- avoid too many overlapping lines.

#### Subplot:

##### What is it?

A subplot in Matplotlib allows you to display multiple plots within a single figure winow. Instead of creating seperate figures for each chart, subplots divide one figure into smaller sections (called axes), where each section can contain its own independent plot --- like line plots, bar plots, scatter plots, etc.

It's extremely useful for comparing different datasets, variables, or visual perspectives side by side.

---
##### Why use it?

Subplots are useful for:

- Comparing multiple visualizations in one figure (e.g., training vs. validation curves).

- Displaying different aspects of the same data (e.g., histogram + line plot + box plot).

- Keeping related plots organized and visually connected.

- Saving space and providing a comprehensive view of multiple trends or metrics together.

---
##### How it works?

A subplot divides the figure into a grid layout of rows and columns. Each “cell” in the grid is an Axes object where you can draw a separate plot.

You can use:

- `plt.subplot()` for simple grids, or

- `plt.subplots()` (recommended) to create multiple axes at once and control them via returned objects.

You can then plot different data on each axis, adjust their titles, labels, and even share axes if needed.

---
##### Syntax

using `plt.subplot()`

````python
plt.subplot(2,1,1)   # (rows, columns, index)
plt.plot([1,2,3], [4,5,6])
plt.title("plot 1")
plt.grid(True)

plt.subplot(2, 1, 2) 
plt.plot([1,2,3], [6,5,4])
plt.title("Plot 2")
plt.grid(True)

plt.tight_layout()
plt.show()
````

using `plt.subplots()` (more modern approach)

````python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 6))

axes[0,0].plot([1,2,3], [4,5,6])
axes[0,0].set_title('Top left')

axes[0,1].bar([1,2,3], [3,2,1])
axes[0,1].set_title("Top right")

axes[1, 0].scatter([1, 2, 3], [2, 3, 4])
axes[1, 0].set_title("Bottom Left")

axes[1, 1].hist([1, 1, 2, 3, 3, 3])
axes[1, 1].set_title("Bottom Right")

plt.tight_layout()
plt.show()
````

##### Parameters

| Parameter          | Description                                                           |
| ------------------ | --------------------------------------------------------------------- |
| `nrows`, `ncols`   | Number of rows and columns of subplots.                               |
| `index`            | Position of the subplot (in `plt.subplot()`).                         |
| `figsize`          | Size of the entire figure (width, height in inches).                  |
| `sharex`, `sharey` | Share x or y axes among subplots for comparison.                      |
| `tight_layout()`   | Adjust spacing automatically to avoid overlap.                        |
| `fig, axes`        | Returned objects from `plt.subplots()` for control and customization. |
---
##### Key Points

- Use `plt.subplots()` for flexible, readable code (preferred over `plt.subplot()`)

- Always call `plt.tight_layout()` to prevent overlapping titles or labels.

- You can loop through axes when plotting similar data across subplots.

- `fig` controls the entire figure, which each `axes` controls one plot.

- Great for multi-metric dashboards or side-by-side analysis.

#### Bar Plot

##### What is it?

A bar plot (or bar chart) is a type of plot in Matplotlib used to represent categorical data with rectangular bars.

Each bar’s height (or length, if horizontal) corresponds to the value of the variable it represents.

It is one of the most common ways to compare quantities across different categories or groups — for example, sales per product, students per class, or visitors per month.

Bar plots can be vertical (`plt.bar()`) or horizontal (`plt.barh()`), and you can also create stacked or grouped bars for comparison.

---
##### Why use it?

Bar plots are useful when you want to:

- Compare discrete categories or groups.

- Visualize magnitude differences quickly and clearly.

- Represent counts, sums, averages, or frequencies across groups.

- Add annotations or color encodings for extra insights.

- Present results that are easy for general audiences to understand.

They are especially effective when dealing with nominal or ordinal variables.

---
##### How it works?

Matplotlib creates bars where:

- The x-axis represents categories (like names or labels).

- The y-axis represents their corresponding values (like counts or measures).

- Each bar starts from a baseline (usually 0) and extends proportionally to its value.

For multiple datasets, you can shift the bars slightly to create grouped or stacked bar charts.

---
##### Syntax:

Vartical Bar Chart

```Python
import matplotlib.pyplot as plt

plt.bar(x, height, color="skyblue", width=0.6, edgecolor="black")
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Bar Plot')
plt.show()
```



Horizontal Bar Chart

```Python
plt.barh(y, width, color='orange', height=0.5)
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')
plt.show()
```



Grouped Bar Plot Example

```Python
import numpy as np

x = np.arange(3)
values1 = [10, 15, 20]
values2 = [8, 12, 18]

plt.bar(x - 0.2, values1, width=0.4, label='2024')
plt.bar(x + 0.2, values2, width=0.4, label='2025')
plt.xticks(x, ['A', 'B', 'C'])
plt.legend()
plt.title('Grouped Bar Plot')
plt.show()
```
---
##### Parameters

| Parameter     | Description                                                                  |
| ------------- | ---------------------------------------------------------------------------- |
| `x`           | Categories (array-like). Positions on the x-axis where bars are drawn.       |
| `height`      | Values (array-like). Heights of the bars.                                    |
| `width`       | Width of bars (default: `0.8`). Smaller = more space between bars.           |
| `bottom`      | The y-coordinate(s) of the bar bases (used for stacking).                    |
| `color`       | Single color or list of colors for each bar (`'blue'`, `'#FF5733'`, etc.).   |
| `edgecolor`   | Color of bar borders (outline).                                              |
| `linewidth`   | Thickness of bar edges.                                                      |
| `align`       | `'center'` (default) or `'edge'` — controls how bars align with x positions. |
| `tick_label`  | Labels for bars (if different from x values).                                |
| `label`       | Legend label for a set of bars.                                              |
| `alpha`       | Transparency (0 = transparent, 1 = opaque).                                  |
| `orientation` | `'vertical'` or `'horizontal'` (for barh).                                   |
| `hatch`       | Pattern filling inside bars (e.g., `'/'`, `'\\'`, `'x'`, `'-'`).             |
| `zorder`      | Layer order (useful when combining with line plots).                         |

Additional functions for variations:

- `plt.barh()` -> Horizontal bars.

- `plt.bar_label()` -> Add value labels to bars.

- `plt.bar_container()` -> Manage multiple bar sets (e.g., in grouped plots).

##### Key Points
- Bar plots are best for categorical comparison, not for continuous trends.

- Use consistent colors for clarity and contrasting colors for emphasis.

- Add data labels (`plt.bar_label()`) for easier reading.

- For multiple datasets:

- Grouped bars → compare categories across groups.

- Stacked bars → show total and contribution per category.

- Use `plt.barh()` when category names are long — it’s easier to read.

- Combine with `plt.subplots()` for dashboards or comparative views.

- Always label axes and include legends when comparing multiple datasets.

#### Histogram

##### What is it?

A histogram is a type of plot in Matplotlib that displays the distribution of numerical data by dividing the data into bins (intervals) and showing how many data points fall into each bin as bars.

It helps visualize the shape, spread, and central tendency of a dataset — like whether data is skewed, normally distributed, or has outliers.

Each bar in a histogram represents a range of values (bin) rather than a single category (unlike a bar plot). The height of the bar shows how many data points fall within that interval.

---
##### Why use it?
Histograms are used to:

- Understand data distribution (normal, uniform, skewed, bimodal, etc.).

- Detect outliers and data spread.

- Choose appropriate statistical models or transformations.

- Visualize frequency, density, or probability distributions.

- Compare distributions between different datasets (with transparency or side-by-side plots).

They are widely used in data analysis, machine learning preprocessing, and statistical visualization.

---
##### How it works?

When you call `plt.hist()`, Matplotlib:

1. Takes a sequence of numerical data.

2. Divides the data range into equal-width bins.

3. Counts how many data points fall into each bin.

4. Plots those counts as vertical bars.

You can control:

- Number of bins (granularity).

- Range of values to include.

- Whether bars represent raw counts, densities, or probabilities.

- Appearence (colors, alignment, transperancy).

---
##### Syntax

```Python
import matplotlib.pyplot as plt

plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.xlable('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```
---
##### Parameters

| Parameter         | Description                                                                     |
| ----------------- | ------------------------------------------------------------------------------- |
| **`x`**           | Input data (array, list, or Series).                                            |
| **`bins`**        | Number of bins or sequence defining bin edges (default = 10).                   |
| **`range`**       | Tuple `(min, max)` specifying the range of values to include.                   |
| **`density`**     | If `True`, the area under the histogram sums to 1 (probability density).        |
| **`color`**       | Color of bars.                                                                  |
| **`edgecolor`**   | Color of bar edges. Useful for visibility.                                      |
| **`histtype`**    | `'bar'` (default), `'step'`, `'stepfilled'`, `'barstacked'`.                    |
| **`alpha`**       | Transparency level (0 = transparent, 1 = solid).                                |
| **`label`**       | Label for the dataset (used with legend).                                       |
| **`orientation`** | `'vertical'` or `'horizontal'`.                                                 |
| **`rwidth`**      | Relative width of bars (0 to 1). 0.9 leaves small gaps between bars.            |
| **`weights`**     | Array of weights (for weighted histograms).                                     |
| **`stacked`**     | If `True`, stacks multiple histograms on top of each other.                     |
| **`align`**       | `'mid'` (default) aligns bars by center; `'left'` or `'right'` shifts position. |
| **`log`**         | If `True`, use log scale for y-axis (useful for skewed data).                   |
| **`label`**       | Label for the dataset, useful for comparison.                                   |
---
##### Illustrated Idea

Example 1: Basic Histogram

````python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)   # 1000 normally distributed values

plt.hist(data, bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
````

Example 2: Comparing Two Distributions

````python
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(2, 1, 1000)

plt.hist(data1, bins=20, alpha=0.6, label='Mean 0', color='skyblue', edgecolor='black')
plt.hist(data2, bins=20, alpha=0.6, label='Mean 2', color='salmon', edgecolor='black')
plt.title('Compating Two Distributions')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

Overlapping histograms show how two datasets differ in mean and spread.

Example 3: Probability Density

````python
data = np.random.randn(1000)

plt.hist(data, bins=20, density=True, color='plum', edgecolor='black')
plt.title('Probability Density Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

##### Key Points

- A histogram is for continuous numerical data; bar plots are for categorical data.

- Adjust bins for clarity:

    - Too few bins → oversimplified.

    - Too many bins → noisy.

- Use `density=True` when comparing to probability distributions (e.g., normal curve).

- Overlay histograms with alpha for comparing multiple distributions.

- Use `plt.grid(axis='y')` to improve readability.

- Use horizontal orientation (`orientation='horizontal'`) for wide datasets.

- Combine with `plt.subplots()` for side-by-side distribution analysis.

#### Scatter Plot

##### What is it?

A scatter plot is a type of visualization in Matplotlib that displays data as Individual points on a two-dimensional plane. Each point represents an observation with two numerical values -- one plotted along the x-axis and the other along the y-axis.

it's ideal for visualizing the relationship or correlation between two variables. For example, height vs weight, study hours vs grades, temperature vs ice cream sales.

---
##### Why use it?

You use scatter plots to:

- Identify relationships or patterns between two variables (positive, negative, or no correlation).

- Detect outliers or unusual data points.

- Visualize clusters or groups in data.

- Compare multiple datasets using different colors, sizes, or markers.

- Understand how one variable changes with respect to another.

Scatter plots are essential for exploratory data analysis (EDA) and regression modeling in data science.

---
##### How it works?

Each point in a scatter plot is defined by two coordinates (x, y).

Matplotlib plots all pairs from two lists, arrays, or series of equal length.

You can customize:

- Color to represent categories or intensity.

- Size to encode another variable.

- Transparency (`alpha`) to handle overlapping points.

- Marker styles to distinguish groups.

---
##### Syntax:

```Python
import matplotlib.pyplot as plt
plt.scatter(x, y, color='blue', marker='o', s=50)
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Scatter plot Example')
plt.show()
```
---
##### Parameters

| Parameter        | Description                                                             |
| ---------------- | ----------------------------------------------------------------------- |
| **`x, y`**       | Input data (lists, arrays, or pandas Series) of equal length.           |
| **`s`**          | Size of points (scalar or array).                                       |
| **`c`**          | Color of points (can be a single color or array for colormap).          |
| **`marker`**     | Shape of marker (`'o'`, `'s'`, `'^'`, `'*'`, `'x'`, etc.).              |
| **`alpha`**      | Transparency (0 to 1). Useful for dense datasets.                       |
| **`edgecolors`** | Color of the marker edges.                                              |
| **`linewidths`** | Width of the marker edges.                                              |
| **`cmap`**       | Colormap name (e.g., `'viridis'`, `'plasma'`) when using a color array. |
| **`label`**      | Label for legend.                                                       |
| **`vmin, vmax`** | Normalize color mapping limits when using numeric color values.         |
---
##### Illustration Idea

Example 1: Basic Scatter Plot

````python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [5,6,7,8,7,9]

plt.scatter(x, y, color='orange', marker='o', s=80)
plt.title('Basic Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
````

Example 2: Colored and Sized by Data

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)     # color varies with data
sizes = 1000 * np.random.rand(50)   # Size varies with data

plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis', edgecolors='black')
plt.title('Scatter plot with varying colors and sizes')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.colorbar(label='Color Scale')
plt.show()
````

Demonstrates how color and size can represent additional variables (useful in 3D-like data representation).

Example 3: Group Comparison

````python
group1_x = [1,2,3,4]
group1_y = [2,3,5,7]

group2_x = [2,3,4,5]
group2_y = [5,6,7,8]

plt.scatter(group1_x, group1_y, color='blue', label='Group 1')
plt.scatter(group2_x, group2_y, color='red', label='Group 2')

plt.title('Comparison of Two Groups')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
````

##### Key Points
- Scatter plots are best for continuous or numerical data.

- Use alpha (transparency) for overlapping points to see dense areas clearly.

- To visualize correlation:

    - Positive correlation: points rise diagonally upward.

    - Negative correlation: points fall diagonally downward.

    - No correlation: random scatter.
    
- Use color (c) and size (s) to represent additional dimensions.

- You can overlay a trendline (using NumPy or Seaborn) to show regression relationships.

- For large datasets, consider using plt.hexbin() or plt.hist2d() for better density visualization.

#### Pie Charts

##### What is it? 

A pie chart is a circular statistical plot in Matplotlib that displays data as slices of a circle, where each slice represents a proportion or percentage of the total.

The entire circle represents 100%, and each wedge (slice) shows the relative size of a category compared to the whole.

Pie charts are often used to visualize categorical data composition — for example, market share, budget allocation, or survey results.

---
##### Why use it?

You use pie charts to:

- Show relative proportions of categories in a dataset.

- Visualize how each part contributes to a whole.

- Make data more intuitive and easy to compare visually.

- Highlight the largest or smallest categories in a distribution.

- Provide a quick visual summary where exact numbers are less important than relative sizes.

However, pie charts are less precise for comparison, so they are best for simple datasets with few categories.

---
##### How it works?

Matplotlib’s plt.pie() divides a circle (360°) into proportional slices.

Each value in the input list or array determines the angle of its corresponding wedge.

You can customize labels, colors, explode slices, add percentage values, and even make the chart look 3D-like by adding shadows or adjusting start angles.

--- 
##### Syntax

```Python
import matplotlib.pyplot as plt

plt.pie(data, labels=labels, autopct="%1.1f%%", startangel=90)
plt.title("Pie Chart Example")
plt.show()
```
---
##### Parameters

| Parameter           | Description                                                                     |
| ------------------- | ------------------------------------------------------------------------------- |
| **`x`**             | Sequence of values (list, array, or Series) representing category sizes.        |
| **`labels`**        | Names of the categories corresponding to each value.                            |
| **`colors`**        | List of colors for each slice.                                                  |
| **`startangle`**    | Rotates the start of the pie chart by given degrees (default = 0).              |
| **`explode`**       | List of values specifying how far each slice should “pull out” from the center. |
| **`autopct`**       | String or function to format percentage display (e.g., `'%1.1f%%'`).            |
| **`shadow`**        | If `True`, adds a shadow for a 3D look.                                         |
| **`counterclock`**  | If `True` (default), slices are drawn counterclockwise.                         |
| **`radius`**        | Radius of the pie chart (default = 1).                                          |
| **`pctdistance`**   | Distance of the percentage text from the center.                                |
| **`labeldistance`** | Distance of the labels from the center.                                         |
| **`textprops`**     | Dictionary of text style properties (e.g., `{'fontsize': 12}`).                 |
| **`wedgeprops`**    | Dictionary to customize wedge appearance (e.g., edge color or width).           |
| **`normalize`**     | If `True`, normalizes input values to sum to 1.                                 |
| **`frame`**         | If `True`, draws a frame around the pie chart.                                  |
---
##### Illustrate Idea

Example 1: Basic Pie Chart

````python
import matplotlib.pyplot as plt

sizes = [40, 25, 20, 15]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']

plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title('Fruit Market share')
plt.show()
````

Example 2: Exploded Slice

````python
sizes = [40, 25, 20, 15]
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
explode = [0.1, 0, 0, 0]

plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%', shadow=True, startangle=90, colors=['gold', 'skyblue', 'lightgreen', 'pink'])
plt.title('Exploded Pie Chart Example')
plt.show()
````

Example 3: Pie Chart with Equal Aspect Ratio

````python
sizes = [30, 40, 20, 10]
labels = ['A', 'B', 'C', 'D']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')    # Insures the pie is drawn as complete circle
plt.title('Equal Aspect Ratio Pie Chart')
plt.show()
````

Example 4: Nested (Donut) Pie Chart

````python
sizes = [50, 30, 20]
labels = ['python', 'java', 'c++']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Add a white circle to make it a donut
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Programming Language Popularity')
plt.show()
````

Example 4: Adding Legends and Colors

````python
sizes = [45, 30, 15, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.legend(labels, title='Cateories', loc='best')
plt.axis('equal')
plt.title('Pie chart with legend')
plt.show()
````

##### Key Points

- Pie charts show part-to-whole relationships, not exact comparisons.

- Best for few categories (4–6 max); too many slices make it unreadable.

- Always use `plt.axis('equal')` to maintain circular shape.

- Use `explode` to emphasize important slices.

- Prefer `autopct='%1.1f%%'` to show percentages.

- Avoid 3D pie charts unless for visual appeal — they distort perception.

- For better precision or many categories, use bar plots instead.

#### Box Plot

##### What is it?

A box plot (also called a box-and-whisker plot) is a graphical representation of the distribution of numerical data through their five-number summary:

- Minimum

- First Quartile (Q1)

- Median (Q2) 

- Third Quartile (Q3)

- Maximum

It displays data spread, central tendency, and outliers in a compact form. Each box shows where the middle 50% of the data lies (between Q1 and Q3), while the "whiskers" extend to show variability outside upper and lower quartile.

---
##### Why use it?

Box plots are useful when you want to:

- Compare distributions across multiple datasets.

- Identify outliers or extreme values.

- Visualize data spread and symmetry.

- Quickly understand data skewness (whether data leans left/right).

- Summarize large datasets without showing every single data point.

They are common in statistical analysis, data science, and quality control where summarizing data variability is key.

---
##### How it works?

Matplotlib’s `plt.boxplot()` computes quartiles and then plots:

1. Box – from Q1 to Q3 (interquartile range, IQR).

2. Median line – horizontal line inside the box (Q2).

3. Whiskers – lines extending to the lowest and highest data points within 1.5 × IQR.

4. Outliers – individual points beyond the whiskers.

Multiple box plots can be drawn side by side for comparing distributions of different variables or groups.

---
##### Syntax

```Python
import matplotlib.pyplot as plt

plt.boxplot(data)
plt.title('Box plot Example')
plt.xlabel('X-axis label')
plt.ylabel('Values')
plt.show()
```

##### Parameters

| Parameter          | Description                                                             |
| ------------------ | ----------------------------------------------------------------------- |
| **`x`**            | Input data (list, array, or list of arrays for multiple boxes).         |
| **`notch`**        | If `True`, draws a notch around the median (shows confidence interval). |
| **`vert`**         | If `True` (default), boxes are vertical. Set `False` for horizontal.    |
| **`patch_artist`** | If `True`, fills boxes with color.                                      |
| **`widths`**       | Width of each box (default = automatic).                                |
| **`meanline`**     | If `True`, shows the mean as a line instead of a point.                 |
| **`showmeans`**    | If `True`, displays the mean as a point or line.                        |
| **`showfliers`**   | If `False`, hides outliers.                                             |
| **`boxprops`**     | Dictionary to style the box (e.g., color, edgewidth).                   |
| **`whiskerprops`** | Style properties for whiskers.                                          |
| **`capprops`**     | Style for caps at the ends of whiskers.                                 |
| **`medianprops`**  | Style for the median line.                                              |
| **`meanprops`**    | Style for mean indicator.                                               |
| **`flierprops`**   | Style for outliers (markers, color, size).                              |
| **`labels`**       | Labels for boxes (useful when comparing multiple groups).               |
---
##### Illustration Idea

Example 1: Basic Box Plot

````python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)

plt.boxplot(data)
plt.title('Basic Box Plot')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
````

Example 2: Multiple Box Plots

````python
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.normal(0, 1, 100),
        np.random.normal(1, 1.5, 100),
        np.random.normal(2, 0.5, 100)]

plt.boxplot(data, tick_labels=['Group 1', 'Group 2', 'Group 3'], patch_artist=True, boxprops=dict(facecolor='skyblue', color='navy'),
medianprops=dict(color='red', linewidth=2))
plt.title('Comparison of Three Groups')
plt.ylabel('Values')
plt.show()
````

Example 3: Horizontal Box Plot

````python
data = np.random.randn(100)

plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen', color='green'))
plt.title('Horizontal Box plot')
plt.xlabel('Values')
plt.show()
````

Example 4: Showing Mean and Hiding Outliers

````python
data = np.random.normal(0, 1, 100)

plt.boxplot(data, showmeans=True, meanline=True, showfliers=False,patch_artist=True, boxprops=dict(facecolor='skyblue', color='blue'),medianprops=dict(color='blue'), meanprops=dict(color='red'))
plt.title('Box Plot with Mean and Hidden Outliers')
plt.show()
````

##### Key Points
- Displays summary statistics in a compact way — perfect for comparisons.

- Outliers are plotted individually, helping identify unusual data points.

- Use `patch_artist=True` to color boxes for clarity.

- Set `vert=False` for horizontal box plots.

- Use `labels` to make comparisons between groups meaningful.

- Use `showmeans=True` or meanline=True to visualize mean values.

- Avoid using box plots for small datasets — summary may be misleading.

- Combine with violin plots to show both summary and distribution shape.

##### Bonus Tip: Adding Grid and Styling

````python
data = [np.random.normal(0, 1, 100),
        np.random.normal(2, 1.5, 100),
        np.random.normal(1, 0.8, 100)]

plt.boxplot(data, patch_artist=True,
            boxprops=dict(facecolor='lavender', color='black'),
            medianprops=dict(color='red', linewidth=2),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'))

plt.title('Styled Box Plot')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
````

##### 💡 How to choose which plots to actually learn and use

If you’re learning for data science or AI, focus on:

1. Line plots → Trends and loss curves

2. Scatter plots → Correlations

3. Histograms → Data distribution

4. Box plots → Data spread and outliers

5. Bar plots → Group comparisons

6. Heatmaps → Correlation matrices or feature maps

Then later (optional, for richer visuals):

- Violin plots (for distribution shape)

- Pair plots / joint plots (Seaborn)

- 3D or contour plots (advanced visualization)

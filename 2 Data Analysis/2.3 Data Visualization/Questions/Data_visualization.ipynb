```python
# Asignment 1: Basic Plotting

#* 1. Create a simple line plot using Matplotlib showing the values of y = 2x + 3 for x ranging from 0 to 10. Label the x-axis as 'X Values' and y axis as 'Y Values'

#* 2. Modify the plot by adding a title 'Linear Function Plot', and change the linestyle to dashed ('--') and marker to ('o')

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

x = np.arange(0, 10)
y = (2*x) + 3
print(x)
print(y)

plt.plot(x, y, marker='o', linestyle="--", color='blue', markersize=8, linewidth=2)
plt.title('Linear Function Plot')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.grid(alpha=0.6)
plt.show()
```

```python
# Assingment 2: Scatter Plot

#* 1. Generate 50 random x-values and 50 random y-values using NumPy. Create a scatter plot using Matplotlib with these values. Label the axes.

#* 2. Customize the scatter plot by increasing the marker size to 80 and adding gridlines with plt.grid(True)

x = np.random.randint(1, 50, (50,))
y = np.random.randint(1, 50, (50,))
c = np.random.randint(1, 50, (50,))

plt.scatter(x, y, c=c,  cmap='viridis', alpha=0.7, s=80, marker='*')
plt.grid(alpha=0.4)
plt.title('Simple Scatter Plot')
plt.xlabel('Random x values')
plt.ylabel('Random y values')
plt.show()
```

```python
# Assignment 3 

#* 1. Create a bar chart of 5 categories: ['A', 'B', 'C', 'D', 'E'] with random integer values. Label the axes and give the chart a title.

x = ['A', 'B', 'C', 'D', 'E']
values = [10, 20, 32, 12, 45]

plt.bar(x, values, color='teal', edgecolor='black', hatch='//')
plt.grid(axis='y', alpha=0.6, linestyle='--')
plt.title('Categories by Values')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
# Assignment 4

#* 1. Create a histogram of 200 random numbers drawn from a normal distribution (mean=0, std=1). Set the number of bins to 20

#* 2. Add a vertical lines at mean and median of the generated data on the same histogram plot

numbers = np.random.randint(1, 50, 200)

mean_val = np.mean(numbers)
median_val = np.median(numbers)

plt.hist(numbers, bins=20, edgecolor='black', color='pink')
plt.axvline(mean_val, color='red', label='Mean')
plt.axvline(median_val, color='blue', label='Median')
plt.legend()
plt.title('Simple Histogram Plot')
plt.xlabel('Range')
plt.ylabel('Frequncy')
plt.grid(axis='y', linestyle='-.', alpha=0.6)
```

```python
# Assignment 5

#* 1. Create a 1x2 subplot figure. In the first subplot, draw a line plot of sin(x) for x from o to 2pi

#* In the second subplot, draw a line plot of cos(x) for the same x range. Add titles to each subplot.

x = np.linspace(0, np.pi*2, 100)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 6))

axes[0].plot(x, np.sin(x), marker='*', color='pink', linestyle='--')
axes[0].set_title('Plot 1')
axes[0].set_xlabel('X')
axes[0].set_ylabel('y')
plt.xlabel('X')
plt.ylabel('Y')

axes[1].plot(x, np.cos(x), marker='o', color='darkblue', linestyle='-.')
axes[1].set_title('Plot 2')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

plt.tight_layout()
plt.show()
```

```python
# Assignment 6: Pie Chart (Matplotlib)

#* 1. Create a pie chart using values [30, 20, 25, 25] and labels ['A', 'B', 'C', 'D']

x = [30, 20, 25, 25]
labels = ['A', 'B', 'C', 'D']

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']
explode = [0, 0.2, 0, 0]

plt.pie(x, labels=labels, colors=colors, explode=explode, shadow=True, autopct='%1.01f%%', startangle=90 )
plt.legend()
plt.title('Simple Pie chart')
```

```python
# Assignment 6: Box Plot (Matplotlib)

#* 1.Generate 100 random values from three different normal distributions and create a boxplot for them.

R1 = np.random.normal(0, 1, 100)
R2 = np.random.normal(0, 1.5, 100)
R3 = np.random.normal(-1, 0.5, 100)

colors=['red', 'green', 'blue']

plt.boxplot(x=[R1, R2, R3], tick_labels=['R1', 'R2', 'R3'], patch_artist=True, boxprops=dict(facecolor='skyblue', color='Navy'))
```

```python
# Assignemt 8: Basic Seaborn Plot (Lineplot)

#* 1. Create a DataFrame with columns 'day' having values 1-30 and 'sales' filled with random integers between 50-200. Use Seaborn to draw a lineplot of day vs sales.

#* 2. Add markers to the seaborn lineplot and apply a dark style using sns.set_style('darkgrid')

df = pd.DataFrame({
    'day': np.arange(1, 31),
    'sales': np.random.randint(50, 200, (30,))
})

sns.lineplot(x=df['day'], y=df['sales'], data=df, marker='o')
sns.set_style('whitegrid')
plt.title('Sales Per Day')
plt.show()
```

```python
# Assignment 9: Seaborn Scatterplot

#* 1. Create a DataFrame with 100 rows containing height and weight columns (random values). Draw a Seaborn scatterplot between height and weight.

#* 2. Add a regression line using sns.regplot() using the same data.

df = pd.DataFrame({
    'height': np.random.randint(100, 180, (100,)),
    'weight': np.random.randint(20, 100, (100,))
})

sns.scatterplot(x=df['height'], y=df['weight'], data=df, legend=['Height', 'Weight'], alpha=0.7)
sns.regplot(x=df['height'], y=df['weight'], data=df, color='red')
plt.title('Height vs Weight')
plt.show()
```

```python
# Assignment 10: Seaborn Barplot

#* 1. Create a DataFrame with columns 'Category' = ['A', 'B', 'C', 'D'] repeated 25 times and 'Value' containing random integers. Use seaborn barplot to show mean value per category.

df = pd.DataFrame({
    'Category': np.random.choice(['A', 'B','C','D'], 100),
    'Value': np.random.randint(1, 100, (100))
})

sns.barplot(x=df['Category'], y=df['Value'], data=df, color='teal', estimator='mean', palette='Blues')
plt.xticks(rotation=45)
sns.set_style('whitegrid')
plt.xticks(rotation=45)
```

```python
# Assignment 11

#* 1.Create a DataFrame with a column 'Gender' containing 200 random values chosen from ['Male', 'Female']. Plot a Seaborn Countplot.

df = pd.DataFrame({
    'Gender':np.random.choice(['Male', 'Female'], 200)
})

sns.countplot(data=df, x='Gender', hue='Gender', palette='muted', legend=True, order=['Male', 'Female'])
plt.title('Count of Male and Female in Gender')
sns.set_theme(style='whitegrid')
plt.xticks(rotation=45)
plt.show()
```

```python
# Assignment 12: Seaborn Histogram/ Displot

#* 1. Create 300 random values from a normal distribution and plot a Seaborn histplot (sns.histplot) with KDE enabled.

#* 2. Change the number of bins to 30 and change the color of the plot using a Seaborn color palette.

data = np.random.normal(0, 1, 300)

sns.histplot(data, bins=30,  palette='muted', kde=True, edgecolor='black')
plt.title('Normal Distribution Using Histplot')
plt.xlabel('Range')
plt.xticks(rotation=45)
```

```python
# Assignment 13: Seaborn Boxplot

#* 1. Create a DataFrame wuth two columns 'Department':['IT', 'HR', 'Finance'] repeated 40 times and 'Salary' containing random integers between 30k-80k. Plot a boxplot of Salary vs Department.

df = pd.DataFrame({
    'Department': np.random.choice(['IT', 'HR', 'Finance'], 40),
    'Salary': np.random.randint(30000, 80000, (40))  
})

sns.boxplot(data=df, x='Department', y='Salary', hue='Department', palette='coolwarm')
sns.swarmplot(data=df, x='Department', y='Salary', color='red')
plt.title('Department vs Salary')
```

```python
# Assignment 14: Pairplot (Seaborn)

#* 1. Create a DataFrame with 3 numerical columns A, B, C with 100 values each. Create a pairplot for all three.

df = pd.DataFrame({
    'A': np.random.randint(50, 100, (100)),
    'B': np.random.randint(50, 100, (100)),
    'C': np.random.randint(50, 100, (100)),
    'Group': np.random.choice(['G1', 'G2', 'G3'], 100)
})


sns.pairplot(data=df, plot_kws={'alpha':0.7, 's':60, 'edgecolor':'w'}, hue='Group', palette='coolwarm')
plt.suptitle('Simple Pairplot', y=1.02)
plt.show()
```

```python
# Assignment 15: Combined Matplotlib + Seaborn

#* 1. Create a DataFrame with 3 columns 'x', 'y', 'category'. Fill x and y with 100 random numbers and assign category randomly to ['A', 'B', 'C']. Use seaborn to create a scatterplot with hue category.

#* 2. On the same figure, overlay a Matplotlib lineplot showing the mean trend line: sort by x and plot mean y values
df = pd.DataFrame({
    'x': np.random.randint(50, 100, (100)), 
    'y': np.random.randint(50, 100, (100)), 
    'category': np.random.choice(['A', 'B', 'C'], 100)
})

mean_line = df.groupby('x')['y'].mean().sort_index()

sns.scatterplot(data=df, x='x', y='y', hue='category', palette='muted',  style='category', legend='auto')
plt.plot(mean_line, color='pink')
plt.title('Simple combined plotting')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

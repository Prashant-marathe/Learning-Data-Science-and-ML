##### Assignment 12: DataFrame Filtering

1. Create a Pandas DataFrame with one column named `Marks`. Filter rows where marks are greate than 50.

2. Filter rows where marks are less than or equal to 50.

```python
df = pd.DataFrame(np.random.randint(1, 100, (20,)), columns=['Marks'])

# Filter rows where marks > 50
print('Marks Greater than 50: ')
print(df[df["Marks"] > 50])
print("-----------------------")

# Filter rows where marks <= 50
print('Marks Greater than or equal to 50: ')
print(df[df['Marks'] <= 50])
```

##### Assignment 14: Basic GroupBy

1. Create a Pandas DataFrame with columns `'Department'` and `'Salary'`. Group the data by department and calculate the mean salary.

2. Count the number of employees in each department.

```python
df = pd.DataFrame({
    "Department": ["IT", "CS", "IT", "Medicine", "CS"],
    "Salary": [32000, 54000, 32800, 54800,48000]
})

# Group by department and mean salary
avg_salary = df.groupby('Department')['Salary'].mean()
print(avg_salary)

# Count the number of employees in each deparment
count =  df.groupby('Department')['Salary'].count()
print(count)
```

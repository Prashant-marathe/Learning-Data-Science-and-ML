#### EDA (Exploratory Data Analysis)
Exploratory Data Analysis (EDA) is an essential first step in any data science project.

It involves taking deep look at the dataset to understand its structure, spot patterns, identify anamolies, and uncover relationships between variables. This process includes generating summary statistics, checking for missing or duplicate data, and creating visualizations like histograms, box plots, and scatter plots. The goal of EDA is to get a clear picture of what the data is telling you before applying any analysis or machine learning models.

By exploring the data throughly, you can make better decisions about how to clean, transform, and model it effictively.

```python
import pandas as pd

df = pd.read_csv('data.csv')

df.head()    # Returns first five rows of DataFrame
df.tail()    # Returns last five rows of DataFrame
df.info()    # print a concise summary of a DataFrame
df.describe()   # Returns mathematical stats for numerical columns
df.columns      # List of column names
df.shape        # Returns shape (rows, columns)
```

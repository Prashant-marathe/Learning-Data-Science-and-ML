```python
# Assignment 6
#* Create a NumPy array of shape (3,3) representing a matrix. Compute its determinant, inverse, and eigenvalues.

arr = np.array([[1,2,3],
                [4,17,6],
                [7,37,9]])

#* 1. Determinant
det = np.linalg.det(arr)
print(f'Determinant: {det}')

#* 2. inverse
inv = np.linalg.inv(arr)
print(f'Inverse:\n {inv}')

#* 3. eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(arr)
print(f'Eigenvalues:\n {eigenvalues}')
print(f'Eigenvectors:\n {eigenvectors}')
```

```python
# Assignment 9
#* Create a structured array with fields 'name' (string), 'age' (integer), 'weight' (float). Add some data and sort the array by age.

student_dtype = np.dtype([
    ('name', 'U10'),
    ('age', 'i4'),
    ('weight', 'f4')
])

structuted_array = np.array([
    ('Alice', 18, 45.0),
    ('Bob', 17, 40),
    ('Charlie', 30, 56.7)
], dtype=student_dtype)

sorted_students = np.sort(structuted_array, order='age')
print(sorted_students)
```

```python
#* Create a structured array with fields 'x' and 'y' (both integers). Add some data and compute the Euclidean distance between each pair of points.

data_dtype = np.dtype([
    ('x', 'i4'),
    ('y', 'i4')
])

data = np.array([
    (1, 7),
    (9, 3), 
    (4, 7), 
    (3, 9)
], dtype=data_dtype)

print(data)


# Extract coordinates as normal 2D NumPy array
coords = np.column_stack((data['x'], data['y']))

# Compute pairwise Euclidean distances
diff = coords[:, np.newaxis, :] - coords[np.newaxis, :, :]
distances = np.sqrt(np.sum(diff**2, axis=-1))

print("\nPairwise Euclidean Distance Matrix:\n", distances)
```

```python
# Assignment 10
#* Create a masked array of shape (4,4) with random integers and mask the elements greater than 10. Compute the sum of the unmasked elements.

data = np.random.randint(0, 20, (4,4))
print(f'Original Array: \n{data}')

#* mask the elements greater than 10
masked_arr = np.ma.masked_greater(data, 10)
print(f'\nMasked Array: \n{masked_arr}')

#* Compute the sum of unmasked elements
sum_unmasked = masked_arr.sum()
print(f'Sum of unmasked elements: \n', sum_unmasked)
```

```python
#* Create a masked array of shape (3, 3) with random integers and mask the diagonal elements. Replace the masked elements with the mean of the unmasked elements.

data = np.random.randint(1, 16, (3,3))
print(f'Original Array: \n {data}')

# mark the diagonal elements
masked_arr = np.ma.array(data, mask=np.eye(3, dtype=bool))
print(f'Masked Array (diagonal masked):\n {masked_arr}')

# Compute the mean of unmasked elements
mean_unmasked = masked_arr.mean()
print(f'Mean of unmasked elements: {mean_unmasked}')

# Replace masked elements with the mean
filled_arr = masked_arr.filled(mean_unmasked)
print(f'Array after replacing maasked (diagonal) elements with mean: \n {filled_arr}')
```

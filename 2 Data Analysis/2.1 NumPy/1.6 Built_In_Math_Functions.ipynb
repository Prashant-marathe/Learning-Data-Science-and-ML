Built in Mathematical Functions in NumPy

---

```python
import numpy as np

arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])

print(arr1)
print(arr2)
```

1. `np.mean()` - Compute the mean (average) of an array.

```python
np.mean(arr1)
```

2. `np.std()` - Compute the standard deviation of an array.

```python
np.std(arr1)
```

3. `np.var()` - Compute the variance of an array.

```python
np.var(arr1)
```

4. `np.min()` - Compute the minimun value of an array.

```python
np.min(arr1)
```

5. `np.max()` - Compute the maximum value of an array.

```python
np.max(arr1)
```

6. `np.sum()` - Compute the sum of all elements in an array.

```python
np.sum(arr1)
```

7. `np.prod()` - Compute the product of all elements in an array.

```python
np.prod(arr1)
```

8. `np.median()` - Compute the median of an array

```python
np.median(arr1)
```

9. `np.percentile()` - Compute the percentile of an array.

```python
np.percentile(arr1, 50)   #* For the 50th percentile (median)
```

10. `np.argmin()` - Returns the index of the minimum value in an array.

```python
np.argmin(arr1)
```

11. `np.argmax()` - Returns the index of the maximum value in an array.

```python
np.argmax(arr1)
```

12. `np.corrcoef()` - Compute the correlation coefficient matrix of two arrays.

```python
np.corrcoef(arr1, arr2)
```

13. `np.unique()` - Find the unique elements in an array.

```python
np.unique(arr1)
```

14. `np.diff()` - Compute the n-th differences of an array.

```python
np.diff(arr1)
```

15. `np.cumsum()` - Compute the cumulative sum of an array.

```python
np.cumsum(arr1)
```

16. `np.log()` - Compute the natural logarithm of an array.

```python
np.log(arr1)
```

17. `np.exp()` - Compute the exponential of an array.

```python
np.exp(arr1)
```

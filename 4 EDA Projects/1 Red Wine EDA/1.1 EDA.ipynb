#### EDA With Red Wine Data

Data Set Information:

These dataset can be viewed as classification or regression tasks. The classes are ordered and not balanced (e.g. there are many more normal wines than excellent or poor ones). Outlier detection can be used to detect the few excellent or poor wines. Also, we are not sure if all input variables are relevant. So it could be interesting to test feature selection methods.

Attribute Information:

Input variable (based on physicochemical tests):

- 1 - fixed acidity

- 2 - Volatile acidity

- 3 - citric acid

- 4 - residual sugar

- 5 - chlorides

- 6 - free sulphur dioxide

- 7 - total sulphur dioxide

- 8 - density

- 9 - pH

- 10 - sulphates

- 11 - alchohol

Output variable (based on sensory data):

- quality (score between 0 to 10)

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt='.2f', linewidths=1, linecolor='white', cmap='YlGnBu')
plt.title('Correlation between variables')
```

```python
df['quality'].value_counts()
```

```python
sns.countplot(x='quality', data=df, hue='quality', legend=False, palette='muted')
plt.grid(axis='y', linestyle="--", alpha=0.4)
plt.title('Quality count of wine')
plt.xlabel('Quality of the wine')
plt.ylabel('Count')
```

```python
df.head()
```

```python
sns.catplot(x='quality', y='alcohol', data=df, kind='box')
```

```python
sns.scatterplot(x='fixed acidity', y='volatile acidity', hue='quality', style='quality', data=df, s=100, markers=True, palette='Set2')
plt.title('Fixed Acidity vs Volatile Acidity')
plt.xlabel('Fixed Acidity')
plt.ylabel('Volatile Acidity')
plt.show()
```

```python
sns.scatterplot(x='free sulfur dioxide', y='total sulfur dioxide', hue='quality', style='quality', data=df, s=100, markers=True, palette='muted')
plt.title('Free Sulfur Dioxide vs Volatile Sulfur Dioxide')
plt.xlabel('Free Sulfur Dioxide')
plt.ylabel('Volatile Sulfur Dioxide')
plt.show()
```

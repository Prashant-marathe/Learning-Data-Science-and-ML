Distance of a line from origin (magnitude)

```python
import numpy as np

# Define an n-dimensional vector A
A = np.array([1,2,3,4,5])

# calculate the Euclidean distance from the origin (L2 norm)
distance = np.linalg.norm(A)

print(f'Euclidean distance from the origin: {distance}')
```

Dot Product (multiplication of 2 vectors)

```python
A = np.array([1,2,3])
B = np.array([4,5,6])

print(np.dot(A, B))
print(A @ B)

# Returns a scalar
```

* dot product (multiplication) follows 2 rules
    - commutative : `A . B = B . A`
    - distributive : `A . (B + C) = A . B + A . C` 

```python
# & check commutative law
A = np.array([1,2,3])
B = np.array([4,5,6])
C = np.array([7,8,9])

# Calculate the dot product 
dot_product_AB = np.dot(A, B)
dot_product_BA = np.dot(B, A)

print(dot_product_AB)
print(dot_product_BA)

# check commutative property
if dot_product_AB == dot_product_BA:
    print("The dot product is commutative")
else:
    print("The dot product is not commutive")

# & Check distributive law
dot_product_A_BC = np.dot(A, B + C)
dot_product_AB_AC = np.dot(A, B) + np.dot(A, C)

print(dot_product_A_BC)
print(dot_product_AB_AC)

if dot_product_A_BC == dot_product_AB_AC:
    print('The dot product is distributive')
else:
    print('The dot product is not distributive')
```

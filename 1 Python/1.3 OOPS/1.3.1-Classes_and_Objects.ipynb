#### Classes and Objects
Object-Oriented Programming (OOP) is a programming paradigm that uses 'objects' to design applications ad computer programs. OOP allows for modeling real-world scenerios using classes and objects.

```python
# Class is an blueprint for creating objects. It has attributes and methods

class Car: #* Declare class Car
    pass

audi = Car()  #* Create objects of the class Car
bmw = Car()

print(audi)
```

```python
# attributes (variables)
audi.windows = 4
print(audi.windows)
```

```python
tata = Car()
tata.doors = 4
print(tata.doors)
```

```python
# Instance variable and methods
class Dog:
    #* constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f'{self.name} Barks!!')

dog = Dog('Browney', 2)
print(dog.name)
print(dog.age)
dog.bark()
```

```python
# Example : Modeling an Bank Account

class BankAccount:
    def __init__(self, owner, balance=0 ):
        self.owner = owner
        self.balance = balance
    
    def deposit(self,amount):
        self.balance += amount
        print(f'The amount has been deposited successfully.\nCurrent balance {self.balance}')

    def withdraw(self, amount):
        self.balance -= amount
        print(f'The amount has been withdrawen successfully.\nCurrent balance is {self.balance}')
    
    def get_balance(self):
        return f'Balance: {self.balance}'
    
account1 = BankAccount('Prince', 2000)
account1.deposit(3000)
account1.withdraw(1000)
account1.get_balance()
```

##### Custom Exception (Raise and Throw Exception) 

```python
class DOBException(Exception):
    pass
```

```python
year = int(input('Enter your birth year: '))
age = 2025 - year

try:
    if age <= 30 and age >= 20:
        print('You fit the age criteria for exam.')
    else:
        raise DOBException
except DOBException:
    print('Your age does not fit age criteria')
except Exception as e:
    print(f'An Error Occured. Reason: {e}')
finally:
    print('This is how you use a custom exception.')
```

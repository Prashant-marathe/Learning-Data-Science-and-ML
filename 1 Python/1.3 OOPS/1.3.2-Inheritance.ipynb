#### Inheritance In Python
Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit attributes and methods from another class.

```python
# Single Inheritance: When a Child class inherits from a single parent class

#* Parent class
class Car:
    def __init__(self, doors, windows, enginetype):
        self.doors = doors
        self.windows = windows
        self.enginetype = enginetype

    def drive(self):
        print(f'The person will drive the {self.enginetype} car.')

#* Child class
class Tesla(Car):
    def __init__(self, doors, windows, enginetype, is_selfdriving):
        super().__init__(doors, windows, enginetype)
        self.is_selfdriving = is_selfdriving

    def selfdriving(self):
        print(f'Tesla supports self driving: {self.is_selfdriving}')


tesla1 = Tesla(5, 4, 'electric', True)
tesla1.selfdriving()
tesla1.drive()
print(tesla1.doors)
print(tesla1.windows)
print(tesla1.enginetype)
```

```python
# Multiple Inheritance: Wnen a class inherits from more than one base class

#* Base class 1
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        print('Subclass must implement this method.')

#* Base class 2
class Pet:
    def __init__(self, owner):
        self.owner = owner

#* Derived class
class Dog(Animal, Pet):
    def __init__(self, name, owner, age):
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
        self.age = age

    def speak(self):
        print(f'{self.name} says woof!!!')

dog = Dog('Browney', 'Prince', 4)
dog.speak()
print(dog.name)
print(dog.owner)
print(dog.age)
```

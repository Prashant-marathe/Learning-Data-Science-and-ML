#### Encapsulation and Abstraction
Encapsulation and abstraction are two fundamental principles of Object-Oriented Programming (OOP) that help in designing robust, maintainable, and reusable code. Encapsulation involves bundling data and methods that operate on the data with a single unit, while abstraction involves hiding complex implementation details and exposing only the necessary features.


##### Encapsulation
Encapsulation is the concept of wrapping data (variables) and methods (functions) together as a single unit. It restricts direct access to some of the object's components, which is a means of preventing accidental interference and misuse of the data.

##### Access Modifiers
Python doesn't have true private variables like Java or c++. But it used naming conventions to indicate how members should be treated.

1. Public: Accessible from anywhere. Syntax-> var
2. Protected: Should be used only within the class and its subclasses (convention only). Syntax-> _var
3. Private: Name mangling makes it hard to access from outside the class. Syntax-> __var


```python
# Example: Basic Encapsulation

class Student:
    def __init__(self, name, marks):
        self.name = name        #* Public variable
        self.__marks = marks    #* Private variable

    def get_marks(self):        #* Getter method
        return self.__marks

    def set_marks(self, marks): #* Setter method
        if 0 <= marks <= 100:
            self.__marks = marks
        else:
            print('Invalid marks!')

#! Create object
s1 = Student('Prashant', 98)

#! Accessinng public variable
print(s1.name)

#! Accessing private variable directly (won't work)
#! print(s1.__marks)  #! âŒAttributeError

#! Access using getter method
print(s1.get_marks())

#! Modify marks using setter
s1.set_marks(99)
print(s1.get_marks())
```

---
Now its not like we can't access private variable outside of the class. We can in Python access private variable outside of the class as just the syntax to access that variable has been changed. Instead of *object.variable* we use *_ClassName__variable.*

```python
print(s1._Student__marks)
```

This shows that Encapsulation is not absolute in Python --- It's convention, not enforcement.

---

```python
# Example: Protected Members

class Car:
    def __init__(self, brand):
        self._brand = brand          #* Protected variable
    
class SportsCar(Car):
    def __init__(self, brand, speed):
        super().__init__(brand)
        self._speed = speed         #* Protected variable
     
    def show(self):
        print(f'Brand: {self._brand}, Speed: {self._speed}')


#! Create object of the SportsCar class
sc = SportsCar('Ferrari', 300)
sc.show()
```

Here _brand and _speed are protected - by convention, they should only be used in subclasses.

```python
# Example with getter and setter methods

class Student:
    def __init__(self, name, age):
        self.__name = name          #* Private variable
        self.__age = age            #* Private variable

    #* getter for accessing private variable __name
    def get_name(self):
        return self.__name
    
    #* Setter was modifying the value of private variable __name
    def set_name(self, name):
        self.__name = name

    #* getter for accessing private variable __age
    def get_age(self):
        return self.__age
    
    #* Setter was modifying the value of private variable __age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else: 
            print('Age cannot be negetive.')


#! Create object of the class Student
s1 = Student('Prince', 21)

#! Display name and age using getter
print(s1.get_name())
print(s1.get_age())

#! Modify name and age via setter
s1.set_name('Prashant')
print(s1.get_name())

s1.set_age(22)
print(s1.get_age())
```

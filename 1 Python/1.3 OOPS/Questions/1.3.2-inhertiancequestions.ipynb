```python
# Assignment 1
#* Create a base class named Animal with attributes name and species. Create a derived class named Dog that inherits from Animal and adds an attribute breed. Create an object of the Dog class and print its attributes.

class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed

dog1 = Dog('Browney', 'BullDog', 'I dont knoe')
print(dog1.name)
print(dog1.species)
print(dog1.breed)
```

```python
# Assignment 2
#* In the Dog class, override the __str__ method to return a string representation of the object. Create an object of the class and print it.

class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed
    
    def __str__(self):
        return f'Name: {self.name}, Species: {self.species}, Breed: {self.breed}'

dog1 = Dog('Browney', 'BullDog', 'I dont knoe')
print(dog1)
```

```python
# Assignment 3
#* In the Dog class, add a method named bark that prints a barking sound. Create an object of the class and call the method.

class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed

    def bark(self):
        return f'{self.name} says Woof!!!'

dog1 = Dog('Browney', 'BullDog', 'I dont knoe')
print(dog1.bark())
```

```python
# Assignment 5
#* In the Athlete class, override the walk method to print a different message. Create a object of the class and call the walk method. Use the super() function to call the walk method of the Walker class

class Walker:
    def walk(self):
        print('I like to walk.')

class Runner:
    def run(self):
        print('I like to run')
    
class Athlete(Walker, Runner):
    def walk(self):
        super().walk()
        print('Athlete likes to walk.')

a1 = Athlete()
a1.walk()
a1.run()
```

```python
# Assignment 6
#* In the Athlete class, add an attribute training_hours and method train that prints the training hours. Create an object of the class and call the method.

class Walker:
    def walk(self):
        print('I like to walk.')

class Runner:
    def run(self):
        print('I like to run')
    
class Athlete(Walker, Runner):
    def __init__(self, training_hours):
        self.training_hours = training_hours
    
    def training(self):
        print(f'Training Hours = {self.training_hours}')

a1 = Athlete(4)
a1.training()
```

```python
# Assignment 7
#* Create a class named A with method show that prints a message. Create two derived classes B and C that inherit from A and override the show method. Create a class D that inherits from both B and C. Create an object of the D class and call the show method. Observe the method resolution order.

class A:
    def show(self):
        print('This is from class A.')
    
class B(A):
    def show(self):
        print('This is overridden method from class B.')

class C(A):
    def show(self):
        print('This is overridden method from class C.')

class D(C, B):
    pass

d1 = D()
d1.show()
```

```python
# Assignment 8
#* Create a base class named Shape with an attribute color. Create a derived class named Circle that inherits from Shape and adds an attribute radius. Use the super() function to initialize the attributes. Create an object of the Circle class and print its attributes.

class Shape:
    def __init__(self, color):
        self.color = color

class Circle(Shape):
    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    
c1 = Circle('red', 4)
print(c1.color)
print(c1.radius)
```

```python
# Assignment 9
#* Create a class named Person with an attribute name. Create a class named Employee with an attribute employee_id. Create an derived class Manager that inherits from both Person and Employee. Use the super() function to initialize the attributes. Create an object of the Manager class and print its attributes.

class Person:
    def __init__(self, name):
        self.name = name

class Employee:
    def __init__(self, employee_id):
        self.employee_id = employee_id

class Manager(Person, Employee):
    def __init__(self, name, employee_id):
        super().__init__(name)
        Employee.__init__(self, employee_id)

m1 = Manager('Prince', 777)
print(m1.name)
print(m1.employee_id)
```

```python
# Assignment 10
#* Create a class named Vehicle with method start that prints a starting message. Create a derived class Car that overrides the start method to print a different message. Use the super() function to call the start method of vehicle class. Create an object of the Car class and call the start method.

class Vehicle:
    def start(self):
        print('Vehicle engine started successfully.')

class Car(Vehicle):
    def start(self):
        print('Car engine started successfully')
        super().start()

c1 = Car()
c1.start()
```

```python
# Assignment 12
#* Create a class named Base1 with an attribute a. Create a class named Base2 with an attribute b. Create a class named Derived that inherits from both Base1 and Base2 and adds an attribute c. Initialize all attributes using the super() function. Create an object of the Derived class and print its attributes.

class Base1:
    def __init__(self, a):
        self.a = a

class Base2:
    def __init__(self, b):
        self.b = b

class Derived(Base1, Base2):
    def __init__(self, a, b, c):
        super().__init__(a)
        Base2.__init__(self, b)
        self.c = c

d1 = Derived(1,2,3)
print(d1.a)
print(d1.b)
print(d1.c)
```

```python
# Assignment 15
#* Create a base class named Device with an attribute brand. Create a derived class Phone that inherits from Device and adds an attribute model. Create another base class Camera with an attribute resolution. Create a derived class Smartphone that inherits from both Phone and Camera. Create an object of the Smartphone class and print its attributes.

class Device:
    def __init__(self, brand):
        self.brand = brand

class Phone(Device):
    def __init__(self, brand, model):
        super().__init__(brand)
        self.model = model 

class Camera:
    def __init__(self, resolution):
        self.resolution = resolution

class SmartPhone(Phone, Camera):
    def __init__(self, brand, model, resolution):
        super().__init__(brand, model)
        Camera.__init__(self, resolution)

s1 = SmartPhone('Samsung', 'Z Fold', '84px')
print(s1.brand)
print(s1.model)
print(s1.resolution)
```

```python
# Assignment 1
#* Write a function that reads the contents of a file named sample.txt and print each line.

def read_file(file_path):
    with open(file_path) as file:
        content = file.readlines()
    return content

file_path = 'sample.txt'
content = read_file(file_path)
for line in content:
    print(line.strip())
```

```python
# Assignment 2
#* Write a function that writes a list of strings to a file named output.txt, with each string on a new line.

str_list = ['line no 1\n', 'line no 2\n', 'line no 3\n', 'line no 4\n', 'line no 5\n']

def write_file(file_path):
    with open(file_path, 'w+') as file:
        file.writelines(str_list)
        file.seek(0)
        content = file.read()
    return content

file_path = 'output.txt'
content = write_file(file_path)
print(content)
```

```python
# Assignment 3
#* Write a function that copies the contents of a file named source.txt to a new file named destination.txt

def copy_file(source_path, destination_path):
    with open(source_path) as source_file:
        content = source_file.read()

    with open(destination_path, 'w') as destination_file:
        destination_file.write(content)

source_path = 'source.txt'
destination_path = 'destination.txt'
copy_file(source_path, destination_path)
```

```python
# Assignment 4
#* Write a function that appends a given string to the end of a file named log.txt

def append_file(file_path, content):
    with open(file_path, 'a') as file:
        file.writelines(content)

file_path = 'log.txt'
content = ['\nline no 1\n', 'line no 2\n', 'line no 3\n', 'line no 4\n', 'line no 5\n']
append_file(file_path, content)
```

```python
# Assignment 5
#* Write a function that reads the contents of a file named document.txt and returns the number of words in the file.

def count_words(file_path):
    with open(file_path) as file:
        lines = file.readlines()
        word_count = sum(len(line.split()) for line in lines)
    return word_count

file_path = 'document.txt'
word_count = count_words(file_path)
print(f"The Word Count of the file is '{word_count}'.")
    
```

```python
# Assignment 6
#* Write a function that finds and replaces all occurences of a given word in a file named data.txt with another word.

def replace_word_in_file(file_path, old_word, new_word):
    with open(file_path) as file:
        content = file.read()
        new_content = content.replace(old_word, new_word)
    
    with open(file_path, 'w') as file:
        file.write(new_content)

file_path = 'data.txt'
old_word = 'Prashant'
new_word = 'Prince'
replace_word_in_file(file_path, old_word, new_word)
```

```python
# Assignment 7
#* Write a function that reads the contents of a file named reverse.txt and prints each line in reverse order.

def read_reverse(file_path):
    with open(file_path) as file:
        lines = file.readlines()
        for line in lines:
            print(line[::-1])

read_reverse('data.txt')
```

```python
# Assignment 8
#* Write a function that reads the contents of a file named stats.txt and returns the number of lines, words, and characters in the file.

def count_file_contents(file_path):
    with open(file_path) as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

lines, words, characters = count_file_contents('data.txt')
print(f"Lines: {lines}, Words: {words}, Characters: {characters}")
```

```python
# Assignment 9
#* Write a function that merges the contents of multiple files into a single file named merged.txt

def merge_files(file_list):
    for file in file_list:
        with open(file) as f:
            content = f.read()
        with open('merged.txt', 'a') as f:
            f.write(content)

file_list = ['data.txt', 'destination.txt', 'document.txt', 'log.txt']
merge_files(file_list)
```

```python
# Assignment 10
#* Write a function that splits a large file named large.txt into smaller files of 100 lines each.

def split_files(file_name, line_per_file=100):
    with open(file_name, 'r') as file:
        count = 0
        part = 1
        outfile = open(f'part_{part}.txt', 'w')

        for line in file:
            outfile.write(line)
            count += 1

        if count == line_per_file:
            outfile.close()
            part += 1
            outfile = open(f'part_{part}.txt', 'w')
            count = 0

        outfile.close()
    print(f'File "{file_name}" has been split into parts of {line_per_file} line each.')

# Test 
split_files('large.txt')
    
```

```python
# Assignment 11
#* Write a function that creates a log file named activity.log and writes log messages with timestamps.

from datetime import datetime

def write_log(message):
    with open('activity.log', 'a') as log_file:
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        log_file.write(f'[{timestamp}] {message}\n')

# Test
write_log('Program started')
write_log('user logged in')
write_log('Data processing completed')
```

```python
# Assignment 12
#* Write a function that reads a binary file named image.bin and writes its contents to another binary file named copy_image.bin.

def copy_binary_file(source_file, destination_file):
    with open(source_file, 'rb') as src:
        with open(destination_file, 'wb') as dest:
            dest.write(src.read())
    print(f'{source_file} copied successfully to {destination_file}')

# Test
copy_binary_file('image.bin', 'copy_image.bin')
```

```python
# Assignment 13
#* Write a function that reads a CSV file named data.csv and prints its contents as a list of dictionaries.

import csv

def read_csv_as_dict(filename):
    with open(filename, mode='r', newline="") as file:
        reader = csv.DictReader(file)
        data = [row for row in reader]
        print(data)
        return data
    
# Test
read_csv_as_dict('data.csv')
```

```python
# Assignment 14
#* Write a function that reads a JSON file named data.json and prints its contents as a Python dictionary.

import json

def read_json_file(filename):
    with open(filename, 'r') as file:
        data = json.load(file)
        print(data)
        return data
    
# Test
read_json_file('data.json')
```

```python
# Assignment 15
#* Write a function that attempts to read a file named protected.txt and handles any permission errors gracefully by printing an error message.

def read_protected_file():
    try:
        with open('protected.txt', 'r') as file:
            content = file.read()
            print(content)
    except PermissionError:
        print('Error: You do not have permission to read "protected.txt"')
    except FileNotFoundError:
        print("Error: The file 'protected.txt' does not exist.")
    else:
        print('File read successfully.')

read_protected_file()
```

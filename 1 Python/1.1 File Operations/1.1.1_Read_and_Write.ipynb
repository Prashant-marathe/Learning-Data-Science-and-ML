### File Operation - Read and Write Files
File handling is crucial part of any programming language. Python provides built-in-functions and methods to read from and write to files, both text and binary. 

```python
## Read whole file

with open('example.txt') as file:
    content = file.read()
    print(content)
```

```python
# Read a file line by line

with open('example.txt') as file:
    for line in file:
        print(line)
```

```python
# Writing into file (Overwriting)

with open('example.txt', 'w') as file:
    content = '''My name is Prashant Marathe.
I am 20 years old.
I am going to be an exceptional Software Engineer in the future.\n'''
    file.write(content)
```

```python
# Writing into file (without overwriting)

with open('example.txt', 'a') as file:
    content = '''`Show me the heart unfitted by foolish dreams and
i'll show you an happy men, But only in their dream can men be truly free, it's always thus and always thus will be.`'''
    file.write(content)
```

```python
# Writing a list of lines in file

with open('example.txt', 'w') as file:
    lines = ['line no 1\n', 'line no 2\n', 'line no 3\n', 'line no 4\n']
    file.writelines(lines)
```

**Binary Files**

```python
# Writing to a binary file

data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# Reading a binary file

with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

*Some Practicle Questions*

```python
# Read a source text file and write its contents to the destination text file

with open('source.txt') as file:
    content = file.read()

with open('destination.txt', 'w') as file:
    file.write(content)
```

```python
# Read a text file and count the number of lines, words, and characters.

def count(file_path):
    with open(file_path) as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = (sum(len(line.split()) for line in lines))
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'destination.txt'
lines, words, characters = count(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
        
```

```python
# Writing and Reading a file
#* The 'w+' mode in Python is used to open a file for both reading and writing. If the file does not exist, it will be created. If the file already exists, its content trancated (i.e., the file will be overwitten)

with open('source.txt', 'w+') as file:
    content = '''This is the era of AI.
No one wants to be left behind.
This is some sort of rule.'''
    file.write(content)
    file.seek(0)  #* Move the file cursor to the beginning
    print(file.read())
```

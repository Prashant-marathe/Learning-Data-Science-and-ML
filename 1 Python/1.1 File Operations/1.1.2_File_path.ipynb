#### OS Module in Python
The os (Operating System) module in Python provides functions to interact with the operating system -- such as reading and writing files, managing directories, working with environment variables, and running system commands.

It acts as a bridge between Python and your computers OS(Windows, macOS, Linux)

```python
import os
```

```python
# Get Current Working Directory

cwd = os.getcwd()
print(f'Current Working Directory is {cwd}')
```

```python
# Create a new directory

new_directory = 'package'
os.mkdir(new_directory)
```

```python
# Listing all files and directories 

content = os.listdir('.')  #* The dot indicates current location
print(content)
```

```python
# Joining paths

dir_name = 'Folder'
file_name = 'File.txt'

#* To get full path
relative_path = os.path.join(dir_name, file_name)
print(relative_path)

#* To get full path
absolute_path = os.path.join(os.getcwd(), dir_name, file_name)
print(absolute_path)
```

```python
# Check if path exists or not

path = 'example1.txt'
if os.path.exists(path):
    print(f"The file path '{path}' exists.")
else:
    print(f"The file path '{path}' does not exist.")
```

```python
# Check if the path is file or directory

path = 'example.txt'
if os.path.isfile(path):
    print(f"The path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory.")
else:
    print(f"The path '{path}' is neither file nor directory.")
```

```python
# Getting the absolute path from relative path

relative_path = 'example.bin'
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```

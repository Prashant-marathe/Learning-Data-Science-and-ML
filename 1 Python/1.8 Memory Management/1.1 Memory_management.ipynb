##### Memory Management in Python

Memory Management in Python involves a combination of automatic garbage collection, reference counting, and various internal optimizations to efficiently manage memory allocation and deallocation. Understanding these mechanisms can help developers write more efficient and robust applications.

Key concepts in Python Memory Management

1. Memory allocation and deallocation

2. Reference counting

3. Garbage Collection

4. The gc module

---

##### Reference Counting

Reference counting is the primary method Python uses to manage memory. Each object in Python maintains a count of references pointing to it. When the references pointing to it. When the reference count drops to zero, the memory occupied by the object is deallocated.

```python
import sys

a = []
print(sys.getrefcount(a))

# Its 2 because one reference is coming from 'a' itself and another one from getrefcount()
```

```python
b = a
print(sys.getrefcount(b))

# Its 2 because one reference is from 'b' itself' another one is from getrefcount() and 3rd one is from 'a'
```

##### Garbage Collection

Python includes a cyclic garbage collector to handle reference cycles. Reference cycles occur when objects reference each other, preventing their reference counts from reaching zero.

```python
import gc

# enable garbage collection 
gc.enable()
```

```python
# disable the garbage collection 
gc.disable()
```

```python
# No. of unreachable objects 
gc.collect()
```

```python
# Get garbage collection stats
print(gc.get_stats())
```

```python
print(gc.collect())
```

##### Memory Management Best Practices

1. Use local variables: Local variables have a shorter lifespan and are freed sooner than global variables. 

2. Avoid Circular Reference: Circular references can lead to memory leaks if not properly managed. 

3. Use Generators: Generators produce items one at a time and only keep one item in memory at a time, making them memory efficient.

4. Explicit Delete Objects: Use the del statement to delete variables and objects explicitly.

5. Profile Memory Usage: Use memory profiling tools like tracemalloc and memory_profiler to identify memory leaks and optimize memory usage.

```python
import gc

class Myclass:
    def __init__(self, name):
        self.name = name
        print(f'Object {self.name} created.')

    def __del__(self):
        print(f'Object {self.name} deleted.')

#* Create circular reference
obj1 = Myclass('obj1')
obj2 = Myclass('obj2')

obj1.ref = obj2 
obj2.ref = obj1

del obj1
del obj2

# # manually trigger the garbage collection
gc.collect()
```

```python
# Generators
# Allow you to produce one item at a time, using memoey efficiently by only keeping one item in the memory at a time.

def generate_numbers(n):
    for i in range(n):
        yield i

# using the generator
for num in generate_numbers(100000):
    print(num)
    if num > 10:
        break
```

```python
# Profiling memory usage using tracemalloc
import tracemalloc

def create_list():
    return [i for i in range(10000)]

def main():
    tracemalloc.start()

    create_list()

    snapshot = tracemalloc.take_snapshot()
    top_stats = snapshot.statistics('lineno')

    print("[Top 10]")
    for stats in top_stats[:10]:
        print(stats)


main()
```

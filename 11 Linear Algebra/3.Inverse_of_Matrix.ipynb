# Inverse of a matrix:

```python
import numpy as np

A = np.array([[1,3,4],
            [6,4,1],
            [2,2,2]])
B = np.array([5, 16, 11])

print(A)
print(B)


# To calculate A inverse
A_inv = np.linalg.inv(A)
print(A_inv)

# Calculate X from A_inv and B
X = np.dot(A_inv, B)
print('Value of X: ', X)
```

So to calculate the solution for any system of linear equations we can use `np.dot(np.linalg.inv(A), B)`

```python
# To calculate euclidean distance in nD
import numpy as np

# ^ Create 2 n-dimensional vectors
A = np.array([1,2,3,4,5])
B = np.array([6,7,8,9,10])

distance = B - A

# & Compute the euclidean distance
print(f'Euclidean distance: {np.linalg.norm(distance)}')
```

#### Use Case in KNN:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.spatial.distance import euclidean
```

```python
np.random.seed(42)
# Generate 5 random 3D vectors
vectors = np.random.rand(5,3)

# Assign a ranfom class (0 or 1) to each vector
classes = np.random.randint(0,2,5)

# plot the vectors on a 3D matplotlib graph
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for vec, cls in zip(vectors, classes):
    ax.quiver(0,0,0, vec[0], vec[1], vec[2], color=("r" if cls == 1 else "b"))

# Get user input for query point (3D vector)
query_vector = np.array(list(map(float, input("Enter the query point (3D vector) seperated by space: ").split())))

# Calculate the distance from the query vector to the 5 vectors and find the nearest neighbor
distances = [euclidean(query_vector, vec) for vec in vectors]
nearest_neighbor_index = np.argmin(distances)

# Output the class of the nearest neighbor
print('The class of the nearest neighbor is: ', classes[nearest_neighbor_index])

# Plot the query vector with a different color
ax.quiver(0,0,0, query_vector[0], query_vector[1], query_vector[2], color="g")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")
plt.show()
```

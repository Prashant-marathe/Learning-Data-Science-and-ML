##### ðŸ˜š SMOTE (Synthetic Minority Oversampling Technique)

SMOTE is a technniue used in machine learning to address imbalanced datasets where the minority class has significantly fewer instances than the majority class. SMOTE involves generating synthetic instances of the minority class by interpolating between existing instances.

```python
# Create a Dataset
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=1000,  #* Number of data points
                    n_features=2,   #* Number of tables
                    n_clusters_per_class=1, #* Number of groups in each table
                    n_redundant=0,
                    weights=[0.90], #* One class should have 90% more values than the other
                    random_state=12)
```

```python
import pandas as pd

df1 = pd.DataFrame(X, columns=['f1', 'f2'])
df2 = pd.DataFrame(y, columns=['target'])
final_df = pd.concat([df1, df2], axis=1)
final_df.head()
```

```python
final_df.target.value_counts()
```

```python
import matplotlib.pyplot as plt
plt.scatter(final_df['f1'], final_df['f2'], c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE

oversample = SMOTE()
```

```python
X, y = oversample.fit_resample(final_df[['f1', 'f2']], final_df['target'])
```

```python
df1 = pd.DataFrame(X, columns=['f1', 'f2'])
df2 = pd.DataFrame(y, columns=['target'])
oversample_df = pd.concat([df1, df2], axis=1)
```

```python
plt.scatter(oversample_df['f1'], oversample_df['f2'], c=oversample_df['target'])
```

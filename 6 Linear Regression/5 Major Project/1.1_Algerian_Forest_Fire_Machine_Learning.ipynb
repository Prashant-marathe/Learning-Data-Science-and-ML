#### Algerian Forest Fire Dataset

Dataset Information: 

The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northewest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012. The dataset includes 11 attributes and 1 output attribute (class) The 244 instances have been classified into fire (138 instances) and not fire (106 instances) classes.

---

**Attribute Information:**

1. Date: (DD/MM/YYYY) Day, month, ('june' to 'september'), year (2012) Weather data observations.

2. Temp: Temperature noon (temperature max) in Celsius degrees: 22 to 44.

3. RH: Relative Humidity in %: 21 to 90.

4. Ws: Wind Speed in km/h: 6 to 29.

5. Rain: total day in mm: 0 to 16.8 FWI Components.

6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5

7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9

8. Drought Code (DC) index from the FWI system: 7 to 220.4

9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5

10. Buildup index (BUI) index from the FWI system: 1.1 to 68

11. Fire Weather Index (FWI) index: 0 to 31.1

12. Classes: two classes, namely Fire and not Fire.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### 1. Data Cleaning

```python
# missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on the Region from 122th index, we can make a new column based on the Region.

1 : "Bajaia Region Dataset" -> 0

2 : "Sidi-Bel Abes Region Dataset"  -> 1

Add new column with region

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1

df = dataset
```

```python
df.info()
```

```python
# Change datatype of Region to int
df[['Region']] = df[['Region']].astype(int)

df['Region'].dtype
```

```python
# Check for null values again
df[df.isnull().any(axis=1)]
```

```python
# Remove the row 122
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.iloc[[122]]
```

```python
# Let's remove this row (122)
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# There are extra spaces in column names just remove those by renaming the columns

df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
# Change the required columns as integer datatypes
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
# Change the columns with float values to float datatype
object_types = [features for features in df.columns if df[features].dtypes == 'O']
object_types

for i in object_types:
    if i != 'Classes':
        df[i] = df[i].astype(float)

df.info()
```

```python
df.describe()
```

```python
# Save clean dataset
df.to_csv('Algerian_Forest_Fire_cleaned.csv', index=False)
```

##### 2. Exploratory Data Analysis

```python
# Percentage based visualization with piechart for Classes
percentage = df_copy['Classes'].value_counts(normalize=True) * 100
```

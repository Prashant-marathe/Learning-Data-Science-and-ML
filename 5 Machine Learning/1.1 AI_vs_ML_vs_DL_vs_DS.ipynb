### ⚔️ AI vs ML vs DL vs DS

---

#### 1. Artificial Intelligence (AI):

- **Definition**: A broad field where machine are designed to simulate human intelligence.

- **Goal**: Make systems that can think, reason, and act like humans.

- **Scope**: The umbrella term that covers ML, DL, DS and other intelligent systems.

- **Examples**: 

    - chatbots

    - Self-driving cars

---

#### 2. Machine Learning (ML):

- **Definition**: A subset of AI where machines learn from data instead of being explicitly programmed.

- **Goal**: Improve performance automatically through experience.

- **Scope**: Uses algorithms that learn patterns from data.

- **Example**: 

    - Spam email detection.

    - Predicting house prices.

---

#### 3. Deep Learning (DL):

- **Definition**: A subset of machine learning that uses artificial neural networks with many layers.

- **Goal**: Handle very complex problems using large datasets and high computing power.

- **Scope**: Great for unstructured data (images, audio, text).

- **Example**:

    - Face recognition

    - Voice Assistants

---

#### 4. Data Science (DS):

- **Definition**: A multidisciplionary field that focuses on extracting insights from data using statistics, programming, ML, and domain knowledge.

- **Goal**: Understand, Visualize, and make decisions from data.

- **Example**:

    - Analyzing customer trends

    - Building Business Dashboards.
